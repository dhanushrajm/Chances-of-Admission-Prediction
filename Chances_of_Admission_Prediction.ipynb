3. Defining Target (y) and Features (X)

```python
X=df1.iloc[:,:7]
y=df1["Chance of Admit "]
```

```python
X.head()
```

```python
X.shape
```

```python
y.head()
```

```python
y.shape
```

4. Splitting the dataset as train and test data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head(6)
```

5. Defining the ML model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

6. Fit the model with training dataset

```python
model.fit(X_train,y_train) #using training dataset
```

```python
model.intercept_
```

```python
model.coef_
```

7. Predicting the model with test dataset

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

8. Classification report/ Accuracy score

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, accuracy_score, classification_report
```

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
mape = mean_absolute_percentage_error(y_test,y_pred)
mape
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
binary_pred = (y_pred >= 0.5).astype(int)
```

```python
binary_pred
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared
r2 = r2_score(y_test, y_pred)

print(f"R-squared: {r2}")
```
